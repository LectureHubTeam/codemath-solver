````python
# Find solutions was duplicated
import json

# Load the data
data_json = json.load(open("/Users/NghiaKhang/Coding/Projects/codemath-solver/submissions_codemath/info.json", encoding="utf-8"))

# Count occurrences of each problem
from collections import defaultdict

problem_counts = defaultdict(list)
for submission_id, submission_data in data_json.items():
    problem = submission_data["problem"]
    problem_counts[problem].append(submission_id)

# Find problems with more than one submission
duplicates = {problem: ids for problem, ids in problem_counts.items() if len(ids) > 1}

# Print duplicated problems and their submission IDs
for problem, ids in duplicates.items():
    print(f"Problem: {problem}, Submission IDs: {ids}, Count: {len(ids)}")
````

````python
# Rename solution files from id to problem name (make sure only one solution file for each problem)
import os
import shutil

# Paths
submissions_dir = "/Users/NghiaKhang/Coding/Projects/codemath-solver/submissions_codemath_refactored"
submissions_renamed_dir = "/Users/NghiaKhang/Coding/Projects/codemath-solver/submissions_codemath_refactored_renamed"

# Ensure the submissions_renamed_dir directory exists
os.makedirs(submissions_renamed_dir, exist_ok=True)

# Track which problems have already been copied
problems_copied = set()

for submission_id, submission_data in data_json.items():
    problem = submission_data["problem"]
    src_file = os.path.join(submissions_dir, f"{submission_id}.py")
    dst_file = os.path.join(submissions_renamed_dir, f"{problem}.py")
    # Only copy the first occurrence of each problem
    if problem not in problems_copied and os.path.exists(src_file):
        shutil.copyfile(src_file, dst_file)
        problems_copied.add(problem)

print(f"Copied {len(problems_copied)} unique problem solutions to {submissions_renamed_dir}")

````

Fix mermaid diagram

````python
import os
import re

def add_quotes_to_labels(mermaid_code):
    lines = mermaid_code.split('\n')
    updated_lines = []

    for line in lines:
        # Bỏ qua dòng comment hoặc subgraph
        if line.strip().startswith('%%') or line.strip().startswith('subgraph'):
            updated_lines.append(line)
            continue

        # Tìm các node label như A[Label] hoặc B{Label}
        line = re.sub(r'(\w+)\[([^\[\]"]+?)\]', r'\1["\2"]', line)
        line = re.sub(r'(\w+)\{([^\{\}"]+?)\}', r'\1["\2"]', line)

        # Tìm các nhánh điều kiện như -- Có -->, -- Không -->
        line = re.sub(r'--\s*([^-"][^->]*?)\s*-->', r'-- "\1" -->', line)

        updated_lines.append(line)

    return '\n'.join(updated_lines)

def process_markdown_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        content = f.read()

    # Regex để tìm khối ```mermaid ... ```
    pattern = r'```mermaid\n(.*?)```'
    matches = re.findall(pattern, content, flags=re.DOTALL)

    updated = False
    for match in matches:
        updated_code = add_quotes_to_labels(match)
        if updated_code != match:
            content = content.replace(match, updated_code)
            updated = True

    if updated:
        with open(filepath, 'w', encoding='utf-8') as f:
            f.write(content)
        print(f"✅ Updated: {filepath}")

def process_folder(folder_path):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.md'):
                process_markdown_file(os.path.join(root, file))

# 📂 Đường dẫn thư mục markdown chứa các file cần cập nhật
folder_path = '/Users/NghiaKhang/Coding/Projects/codemath-solver/explanations'  # <-- sửa thành thư mục của bạn
process_folder(folder_path)
````
